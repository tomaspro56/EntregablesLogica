```python
import pandas as pd
import numpy as np

# 1.2 Cargar el archivo con el separador ;
# Usamos el nombre exacto de tu archivo|1
df = pd.read_csv('Cars.csv', sep=';')

# 1.3 y 1.4 Primer vistazo
print("Primeros 5 registros")
display(df.head())

print("Últimos 5 registros")
display(df.tail())

# 1.5 y 1.6 Dimensiones e Info
print(f"Dimensiones: {df.shape}")
print("\n--- Información General del Dataset ---")
df.info()
```

```python
# 2.1 Verificar valores nulos por columna
print("Valores Faltantes")
print(df.isnull().sum())

# 2.2 Verificar si 'Fuerza' es un número o texto
print("Tipo de dato de la columna Fuerza")
print(df['Fuerza'].dtype)
```

```python
# Contar cuántas filas tenemos antes de borrar
filas_antes = len(df)

# Borrar todas las filas que tengan AL MENOS un valor vacío (NaN)
df = df.dropna()

# Contar cuántas filas quedaron
filas_despues = len(df)

print(f"Limpieza de nulos")
print(f"Filas eliminadas: {filas_antes - filas_despues}")
print(f"Filas restantes: {filas_despues}")

# Verificar que ya no queden nulos en ninguna columna
print("Conteo de nulos final:")
print(df.isnull().sum())
```
