```python
import pandas as pd
import numpy as np

# 1.2 Cargar el archivo con el separador ;
df = pd.read_csv('Cars.csv', sep=';')

# 1.3 y 1.4 Primer vistazo
print("Primeros 5 registros")
display(df.head(5))

print("Últimos 5 registros")
display(df.tail(5))

# 1.5 y 1.6 Dimensiones e Info
print(f"Dimensiones: {df.shape}")
print("Información General del Dataset")
df.info()
```

```python
# 2.1 Verificar valores nulos por columna
print("Valores Faltantes")
print(df.isnull().sum())

# 2.2 Verificar si 'Fuerza' es un número o texto
print("Tipo de dato de la columna Fuerza")
print(df['Fuerza'].dtype)
```

```python
# Contar cuántas filas tenemos antes de borrar
filas_antes = len(df)

# Borrar todas las filas que tengan AL MENOS un valor vacío (NaN)
df = df.dropna()

# Contar cuántas filas quedaron
filas_despues = len(df)

print(f"Limpieza de nulos")
print(f"Filas eliminadas: {filas_antes - filas_despues}")
print(f"Filas restantes: {filas_despues}")

# Verificar que ya no queden nulos en ninguna columna
print("Conteo de nulos final:")
print(df.isnull().sum())
```

```python
# Contar cuántas filas están exactamente repetidas
n_duplicados = df.duplicated().sum()
print(f"Se encontraron {n_duplicados} registros duplicados.")
```

```python
# 3.1 Obtener el resumen estadístico completo
print("Resumen Estadístico del Dataset")
display(df.describe())
```

```python
# 3.2 Calculando medidas específicas para la columna Fuerza
print(f"Media de Fuerza: {df['Fuerza'].mean():.2f}")
print(f"Mediana de Fuerza: {df['Fuerza'].median():.2f}")
print(f"Moda de Fuerza: {df['Fuerza'].mode()[0]}")
```

```python
# 3.3 Matriz de correlación entre las variables clave
columnas_interes = ['Peso', 'Fuerza', 'Cilindros', 'MPG']
matriz = df[columnas_interes].corr()

print("Matriz de Correlación")
display(matriz)
```

```python
# 4.1 Seleccionar solo algunas columnas específicas
df_reducido = df[['Modelo', 'Fuerza', 'Origen']]
print("Vista del dataset reducido")
display(df_reducido.head())
```

```python
# 4.2 Filtrar autos con Fuerza mayor a 200
autos_potentes = df[df['Fuerza'] > 200]
print(f"Se encontraron {len(autos_potentes)} autos con más de 200 de fuerza.")
display(autos_potentes)
```

```python
# 4.3 Filtro de autos del año 70 Y con fuerza > 200
filtro_pro = df[(df['Ano'] == 70) & (df['Fuerza'] > 200)]

print("Autos más potentes del año 70")
display(filtro_pro)
```

```python
# 4.4 Filtrar solo autos de origen Japonés
autos_japoneses = df[df['Origen'] == 'Japan']
print(f"Cantidad de autos japoneses: {len(autos_japoneses)}")
```

```python
# 5.1 Conteo de autos por Origen
print("Cantidad de autos por Origen")
print(df['Origen'].value_counts())

# Conteo de autos por Categoría
print("Cantidad de autos por Categoría")
print(df['Categoria'].value_counts())
```

```python
# 5.2 Agrupar por Origen y calcular promedios de MPG y Fuerza
agrupado_origen = df.groupby('Origen')[['MPG', 'Fuerza']].mean()

print("Promedios de Rendimiento por Origen")
display(agrupado_origen)

# Agrupar por Categoría para ver el Peso promedio
agrupado_cat = df.groupby('Categoria')[['Peso']].mean()
print("Peso promedio por Categoría")
display(agrupado_cat)
```

Conclusiones
- Limpieza de Datos: Se identificaron y eliminaron filas con valores nulos, reduciendo el dataset de 394 a 372 registros, lo que garantiza cálculos precisos de promedios y tendencias.

- Integridad: Se confirmó que el dataset no contenía registros duplicados.

- Desempeño: Las estadísticas mostraron que variables como la Fuerza y el Peso son determinantes para entender el rendimiento de los autos, permitiendo comparar de forma clara los diferentes orígenes.

- Gracias a comandos como dropna() y drop_duplicates(), logramos depurar el dataset de manera rápida y eficiente, eliminando errores y repeticiones mediante código.
